```python
import tensorflow as tf
import glob

from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#categories = glob.glob('PSNR/*')

datasets = ['resnet_pgd0.1']

dataset = datasets[0]

#adversarial_images = glob.glob('datasets/{}/data/*.jpg'.format(dataset))
adversarial_images = glob.glob('datasets/datasets2/{}/data/*.jpg'.format(dataset))
original_images = glob.glob('datasets/coco_224/data/*.jpg')

print(len(adversarial_images))
print()
print(len(original_images))
```

```python
def plot_images(im1,im2):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(1, 3, 1)
    imgplot = plt.imshow(im1)
    ax.set_title('Original')
    plt.axis('off')
    
    ax = fig.add_subplot(1, 3, 2)
    imgplot = plt.imshow(im2)
    ax.set_title('Perturbed')
    plt.axis('off')
    
    ax = fig.add_subplot(1, 3, 3)
    imgplot = plt.imshow((im2-im1)* 0.5 + 0.5)
    ax.set_title('Perturbed')
    plt.axis('off')
    plt.show()
    

# Read images from file.
psnr_values = []
calculator = 0
for orig, ae in zip(original_images, adversarial_images):
    im1 = tf.io.read_file(orig)
    im2 = tf.io.read_file(ae)
    # Compute PSNR over tf.uint8 Tensors.
    im1 = tf.image.decode_jpeg(im1, channels=3)
    im2 = tf.image.decode_jpeg(im2, channels=3)
    psnr1 = tf.image.psnr(im1, im2, max_val=255)
    
    # Compute PSNR over tf.float32 Tensors.
    im1 = tf.image.convert_image_dtype(im1, tf.float32)
    im2 = tf.image.convert_image_dtype(im2, tf.float32)
    psnr2 = tf.image.psnr(im1, im2, max_val=1.0)
    
    
    calculator = calculator + 1
    print("Image", calculator)
    print(psnr2.numpy())
    psnr_values.append(psnr2.numpy())
    
    #plot_images(im1, im2)
    
    #break
```

```python
average = 0
number_of_images = 5000
for value in psnr_values:
    #print(round(value.astype(float),2))
    average = average + round(value.astype(float),2)
print(dataset, "average PSNR:", average/number_of_images)
```
