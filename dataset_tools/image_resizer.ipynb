```python
# resize coco images into 224x224
from PIL import Image
import glob
import numpy as np
```

```python
list_images = glob.glob('val2017/data/*.jpg')
```

```python
print((list_images[0]))
```

```python
def get_filename(path):
    name = path.split("\\")[1].split('.').pop(0)
    #print(name)
    return name

counter = 0

for file in list_images:
    filename = get_filename(file)
    image = Image.open(file)
    new_image = image.resize((224, 224))
    new_image.save('resized_data/{}.jpg'.format(filename))
    
    counter += 1
    print('Counter:', counter, '\n', 'Original_size:', image.size, '--->', 'New_size:', new_image.size)
    #print(image.size) # Output: (1920, 1280)
    #print(new_image.size) # Output: (400, 400)
    #break
```
